For your own Naive Bayes:
```
1) Split data into train and test
2) Count how often a word appears in the positive training set and negative training set
3) Change that into log probabilities
4) For each review in the test set, simply sum up it's words' log probability for being positive and negative
5) Compare the two sums and assign the class that scores higher
```

````python
import os
import pandas as pd
import re
from collections import defaultdict
import numpy as np
````

````python
path_pos = '/home/ali/ds/metis/metisgh/pair_programming/review_polarity/txt_sentoken/pos/'
path_neg = '/home/ali/ds/metis/metisgh/pair_programming/review_polarity/txt_sentoken/neg/'

reviews_pos = []
reviews_neg = []

for filename in os.listdir(path_pos):
    with open(path_pos + filename, 'r') as content_file:
        review = content_file.read()
        reviews_pos.append(review)
    
for filename in os.listdir(path_neg):
    with open(path_neg + filename, 'r') as content_file:
        review = content_file.read()
        reviews_neg.append(review)
````

````python
for i in range(len(reviews_pos)):
    reviews_pos[i] = re.sub("[^a-zA-Zé]", " ", reviews_pos[i])

for i in range(len(reviews_neg)):
    reviews_neg[i] = re.sub("[^a-zA-Zé]", " ", reviews_neg[i])
    
````

````python
pos_test = reviews_pos[:500]
pos_train = reviews_pos[500:]
neg_test = reviews_neg[:500]
neg_train = reviews_neg[500:]
````

````python
pos_dict = defaultdict(int)
neg_dict = defaultdict(int)

word_count_pos = 0
for x in pos_train:
    words = x.split()
    for word in words:
        pos_dict[word] += 1
        word_count_pos += 1
````

````python
word_count_neg = 0
for x in neg_train:
    words = x.split()
    for word in words:
        neg_dict[word] += 1
        word_count_neg += 1
````

````python
for key, value in pos_dict.items():
    pos_dict[key] = value / float(value + neg_dict[key])
````

````python
for key, value in neg_dict.items():
    neg_dict[key] = value / float(value + pos_dict[key])
````
