```python
import numpy as np
import matplotlib.pyplot as plt
% matplotlib inline
from copy import deepcopy
points = np.random.rand(100,2)
import seaborn
```

```python
def distance(A, B):
    squares = [(a - b) ** 2 for a, b in zip(A, B)]
    return sum(squares) ** 0.5
```

Complete the function below. The centroid is just the average of all points in that cluster:

```python
def updateCentroids(points,centroids,clusters):
    for c in range(0, len(centroids -1)):
        cent_points = points[clusters == c]
        centroids[c] = cent_points.mean(axis=0)
```

Complete the function below. For each point, find the distance to all centroids and assign it to the closest one (the distance function is already available above):

```python
def assignPoints(points,centroids,clusters):
    change = False
    for p in range(0, len(points -1)):
        dist = [0] * len(centroids)
        for i in range(0, len(centroids -1)):
            dist[i] = distance(points[p], centroids[i])
        min_dist = dist.index(min(dist))
        if clusters[p] != min_dist:
            clusters[p] = min_dist
            change = True
    return change
```

**Points** is an array of length 100. **Clusters** is an array of the same length. We will use it to assign each point a cluster number. Clusters[0] will be the cluster that point 0 belongs to. **Centroids** are the center of each cluster. So it's an array of length K. We intialize the centroids to be the first K points.

```python
def KMeans(points,k=3):
    clusters = np.zeros(len(points))
    centroids = deepcopy(points[:k])
    while assignPoints(points,centroids,clusters):
        updateCentroids(points,centroids,clusters)
    plt.scatter([p[0] for p in points],[p[1] for p in points],c=clusters)
    plt.show()
```

```python
KMeans(points, 3)
```
